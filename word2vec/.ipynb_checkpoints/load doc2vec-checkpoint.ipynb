Read 100 sentences from amazon reviews and cluster them into 10 topics

```python
import json 
import gzip 
def parse(path): 
    g = gzip.open(path, 'r') 
    for l in g: 
        yield eval(l)
```

```python
import nltk
import gensim

sent_detector = nltk.data.load('tokenizers/punkt/english.pickle')
review_item_iterator = iter(parse("../../download/reviews_Cell_Phones_and_Accessories.json.gz"))

def sentences():
    
    for review_item in review_item_iterator:
        reviewText = review_item["reviewText"]
        for sentence in sent_detector.tokenize(reviewText):
            preprocess_sentence = gensim.utils.simple_preprocess(sentence)
            yield preprocess_sentence 
        
```

```python
import numpy as np
cluster_sentences = [next(sentences()) for it in range(1000)]
cluster_sentences_vec = [model.infer_vector(s) for s in cluster_sentences]
#print(sentence_vec)
np_array = np.asarray(cluster_sentences_vec)
print(len(np_array))
```

```python
import nltk
from nltk.cluster.kmeans import KMeansClusterer
nc = 10
data = np_array

kclusterer = KMeansClusterer(nc, distance=nltk.cluster.util.cosine_distance, repeats=25)
assigned_clusters = kclusterer.cluster(data, assign_clusters=True)
```

```python
print(assigned_clusters)
np_assigned_clusters = np.asarray(assigned_clusters)
```
