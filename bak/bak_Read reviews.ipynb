Call REST API to read a set of reviews:
https://m.aliexpress.com/ajaxapi/EvaluationSearchAjax.do?type=all&index=1&pageSize=20&productId=32797368004&country=US


```python
import requests
base_url = 'https://m.aliexpress.com/ajaxapi/EvaluationSearchAjax.do?type=all&pageSize=80&country=US'
def importReviews(doc_complete, product_id):
    totalPage = 10000000000000000000
    n = 1
    while (n<=totalPage):
        #print(n)
        url = base_url + "&productId="+str(product_id)+"&index="+str(n)
        print(url)
        resp = requests.get(url)
        if resp.status_code != 200:
            # This means something went wrong.
            raise ApiError('GET /EvaluationSearchAjax/ {}'.format(resp.status_code))
        resp_json = resp.json()
        doc_complete_n = [review["buyerTranslationFeedback"] for review in resp_json["evaViewList"] if "buyerTranslationFeedback" in review.keys()]
        doc_complete= doc_complete + doc_complete_n
        if (n==1):
            #print(resp_json)
            totalPage = int(resp_json["totalPage"])
            print(str(product_id)+" totalPage "+str(totalPage))
        print(n)    
        n=n+1
    return doc_complete
```

```python
product_ids = [32797368004,32660593076,32763320059]
doc_complete = []
for product_id in product_ids:
    doc_complete=importReviews(doc_complete,product_id)

```

```python
import nltk

sentence = "At eight o'clock on Thursday film morning word line test best beautiful Ram Aaron design"

full_text = " ".join([review for review in doc_complete])

tokens = nltk.word_tokenize(full_text)

tagged = nltk.pos_tag(tokens)
print(tagged)
```

```python
#nouns = [token for token, pos in tagged if pos.startswith('N')]
#print(nouns)
  	
grammar = r"""
  STATEMENTS:  {<CD|IN|DT|JJ|PP|RB|\$>*<N.*>+<V.*>*<CD|IN|DT|JJ|PP|RB|\$>*}  
"""
cp = nltk.RegexpParser(grammar)
statements_tagged = cp.parse(tagged)
print(statements_tagged)
#doc_complete = nouns
```

```python
def traverse(t,statements):
    try:
        t.label()
    except AttributeError:
        n=0
    else:
        if t.label() == 'STATEMENTS': 
            statement = " ".join([word for (word,pos) in t.leaves()])
            #print(statement)
            statements.append(statement)
        # Now we know that t.node is defined
        #print('(', t.label(), end=" ")
        for child in t:
            traverse(child,statements)
        #print(')', end=" ")
    return statements

statements=[]
statements=traverse(statements_tagged,statements)
print(statements)
doc_complete = statements
 
```

```python
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string
stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()
def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized

doc_clean = [clean(doc).split() for doc in doc_complete]   
#print(doc_clean)
```

```python
from nltk import FreqDist
#print(doc_clean)
#words = nltk.tokenize.word_tokenize(doc_clean[1])
words = []
for w in [d for d in doc_clean]:
    words = words + w
fdist = FreqDist(statements) #words
most_common_words_freq = fdist.most_common(200)
print(most_common_words_freq)
#fdist.plot()
most_common_words= [w for (w,c) in most_common_words_freq]

def clean(doc):
    only_frqt_words = " ".join([w for w in doc if w in most_common_words])
    return only_frqt_words 

#doc_clean2 = [clean(doc) for doc in doc_clean] 
doc_clean2 = [clean(doc).split() for doc in doc_clean] 
print(doc_clean2)
```

```python
# Importing Gensim
import gensim
from gensim import corpora

# Creating the term dictionary of our courpus, where every unique term is assigned an index. 
dictionary = corpora.Dictionary(doc_clean)

# Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above.
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
```

```python
# Creating the object for LDA model using gensim library
Lda = gensim.models.ldamodel.LdaModel

# Running and Trainign LDA model on the document term matrix.
ldamodel = Lda(doc_term_matrix, num_topics=50, id2word = dictionary, passes=50,eta="auto")
```
